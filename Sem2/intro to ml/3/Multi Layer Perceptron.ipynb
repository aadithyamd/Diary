### Read the data

```python
from sklearn.datasets.samples_generator import make_blobs

(X,Y) =  make_blobs(n_samples=300,n_features=2,centers=4,cluster_std=1.05,random_state=40)
#we need to add 1 to X values (we can say its bias)
y = ((Y == 0) + (Y == 3)).astype(int)
plt.scatter(X[:,0],X[:,1],marker='o',c=y)
#plt.axis([-5,10,-12,-1])
plt.show()
y
```

### Normalize the data

```python
y = y.reshape(-1,1)
n = np.where(y == 0)
y[n,0] = -1

mean = X.mean(axis=0)
std = X.std(axis=0)
X = (X- mean) / std
x = np.insert(X, 0, 1,axis = 1)
#x = x / np.sqrt((x ** 2).sum(axis=1).reshape(-1,1))
```

### Split the data to train & test (70 - 30 split)

```python
d = int(y.shape[0] * 0.7)

xtrain = x[:d]
xtest = x[d:]
ytrain = y[:d]
ytest = y[d:]
```
